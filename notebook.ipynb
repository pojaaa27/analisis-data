## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
fig, ax = plt.subplots(figsize=(14, 5))

ax.plot(range(len(monthly_sales)), monthly_sales['revenue'], 
        marker='o', linewidth=2, color='#1f77b4', markersize=8)
ax.fill_between(range(len(monthly_sales)), monthly_sales['revenue'], alpha=0.3)

ax.set_title('Monthly Revenue Trend 2023', fontsize=16, fontweight='bold')
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Revenue (Rp)', fontsize=12)
ax.set_xticks(range(len(monthly_sales)))
ax.set_xticklabels(monthly_sales['month'], rotation=45)
ax.grid(True, alpha=0.3)

ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'Rp {x/1e6:.1f}M'))

plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
fig, ax = plt.subplots(figsize=(12, 6))

colors = plt.cm.Blues(np.linspace(0.4, 0.9, len(category_sales)))[::-1]
bars = ax.barh(category_sales['category'], category_sales['total_sold'], color=colors)

ax.set_title('Product Categories by Sales Volume', fontsize=16, fontweight='bold')
ax.set_xlabel('Total Items Sold', fontsize=12)
ax.invert_yaxis()

for bar, value in zip(bars, category_sales['total_sold']):
    ax.text(value + 5, bar.get_y() + bar.get_height()/2, 
            f'{value:,}', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```
